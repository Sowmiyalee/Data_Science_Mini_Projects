```python
#import libraries
import pandas as pd
import numpy as np
```

```python
#import data
data = pd.read_csv("/content/book.csv", encoding='latin1')
data
```

**Apriori Algorithm**

## 1. Association rules with 10% Support & 70% confidence lvl

```python
from mlxtend.frequent_patterns import apriori

#With 10% Support
frequent_itemsets = apriori(data,min_support=0.1,use_colnames=True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules

#With 70% Confidence
rules = association_rules(frequent_itemsets, metric='lift',min_threshold=0.7)
rules
```

```python
# A leverage value of 0 indicates independence. Range will be [-1 1]
# +1 indicates +covariance, both increses or both decreses
# -1 indicates -covariance, one increses and other decreses
# 0 indicates two variance are independent of each other
# A high conviction value means that the consequent is highly depending on the antecedent and range [0 inf]
```

```python
rules.sort_values ('lift', ascending=False)
```

```python
# Lift Ratio > 1 is a good influential rule in selecting the associated transactions
rules[rules.lift>1]
```

```python
rules[['support','confidence','lift']].hist()
```

```python
# visualization of obtained rule
import matplotlib.pyplot as plt
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence') 
plt.show()
```

```python
import seaborn as sns
sns.scatterplot('support', 'confidence', data=rules, hue='antecedents')
plt.legend(loc='upper left',bbox_to_anchor = (1,1))
plt.show()
```
