```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
#import data
data = pd.read_csv("/content/wine.csv")
data
```

## EDA

```python
data['Type'].value_counts()
```

```python
data1=data.drop(['Type'],axis=1)
data1
```

```python
data1.shape
```

```python
# Converting to Numpy array
data_ary = data1.values
data_ary
```

```python
#Normalizing to numerical data
data_norm = scale(data1)
data_norm
```

## PCA Implementation - Principle component analysis

```python
#Apply PCA fit transform to dataset
pca = PCA(n_components=13)

data_pca = pca.fit_transform(data_norm)
data_pca
```

```python
#PCA components matrix or covariance matrix
pca.components_
```

```python
#the amount of variance that each PCA has
var = pca.explained_variance_ratio_
var
```

```python
#cummulative variance of each pca
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
#Variance plot for PCA component obtained
plt.plot(var1,color='magenta')
```

```python
#Final dataframe
data_final = pd.concat([data['Type'],pd.DataFrame(data_pca[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
data_final
```

```python
#Visualization of PCA
fig=plt.figure(figsize=(16,12))
sns.scatterplot(data=data_final)
```

## 1. Heirarchical clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
#we already have normalized data
dendrogram = sch.dendrogram(sch.linkage(data_norm,'complete'))
```

```python
dendrogram = sch.dendrogram(sch.linkage(data_norm,'ward'))
plt.title('Dendrogram')
plt.xlabel('wine')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
#create clusters
hclusters = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hclusters
```

```python
#predict cluster in dataset

y = pd.DataFrame(hclusters.fit_predict(data_norm),columns=['clustersid'])
y
```

```python
y['clustersid'].value_counts()
```

```python
#Add cluster to dataset
data_heirarchical= data1.copy()
data_heirarchical['clustersid'] = hclusters.labels_
data_heirarchical
```

## 2. K-Means clustering

```python
from sklearn.cluster import KMeans
```

```python
# we already have normalized data - data_norm
# Use Elbow Graph to find optimum number of  clusters (K value) from K values range
# The K-means algorithm aims to choose centroids that minimise the inertia, or within-cluster sum-of-squares criterion WCSS 
# random state can be anything from 0 to 42, but the same number to be used everytime,so that the results don't change. 
```

```python
#within-cluster sum-of-square criterian
wcss=[]
for i in range (1,11):
  kmeans = KMeans(n_clusters=i, random_state=0)
  kmeans.fit(data_norm)
  wcss.append(kmeans.inertia_)
```

```python
#plot k values range vs wcss to get elbow plot
plt.plot(range(1,11),wcss)
plt.title('Elbow graph')
plt.xlabel('no of cluster')
plt.ylabel('WCSS')
plt.show()
```
