## Model Building

```python
#Since the target variable is continious, we create a groups based on mean
# "Small" <=7.49
# 7.49 => "large"
```

```python
data["sales"] = "small"
data.loc [data["Sales"]>7.49, "sales"] = "large"
#data.drop(["Sales"],axis=1,inplace = True)
data
```

```python
sns.catplot(x='sales', y='Sales', data =data)
```

```python
sns.catplot(x='sales', y='Sales', data =data, kind = 'box')
```

```python
#Dividing the data into independent and dependent variable
X = data.iloc[:,0:11]
y = data.iloc[:,11]
```

```python
#Splitting the dataset into train and test dataset
from sklearn.model_selection import train_test_split
X_train, X_test ,y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=40)
```

```python
y_train.value_counts()
```

### Model building by classifier method - categorical variable

```python
#fitting Random forest classification to the training set
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=5, max_depth = 3, random_state = 1)
classifier.fit(X_train,y_train)
```

```python
#Model prediction
y_pred = classifier.predict(X_test)
y_pred
```

```python
y_test
```

```python
#Model accuracy
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_pred, y_test)
print(cm)

from sklearn.metrics import accuracy_score
acc = accuracy_score(y_pred, y_test)
print("Accuracy=", acc*100)
```

```python
pd.DataFrame(index=X.columns,data=classifier.feature_importances_,columns=['Feature Importance']) 
#Only Sales matters since our tree is based on it. 
```
