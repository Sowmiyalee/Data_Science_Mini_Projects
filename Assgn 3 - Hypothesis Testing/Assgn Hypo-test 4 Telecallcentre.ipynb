# **Question 4**

TeleCall uses 4 centers around the globe to process customer order forms. They audit a certain %  of the customer order forms. Any error in order form renders it defective and has to be reworked before processing.  The manager wants to check whether the defective %  varies by centre. Please analyze the data at 5% significance level and help the manager draw appropriate inferences

We are going to conduct a Test of Independence using Chi-Square test with Contingency table
We need to check whether the mean of any of these samples are different or the same




*  H0:  Independence of categorical variables (defective % does not vary by centers)
*  H1: Dependence of categorical variables (defective % varies by centre and somewhat/significantly related)

```python
centres= pd.read_csv('/content/Costomer+OrderForm.csv')
centres.head()
```

```python
centres.describe()
```

```python
centres.isnull().sum()
```

```python
# Checking data type
centres.info()
```

```python
# Checking Counts

print(centres['Phillippines'].value_counts(), '\n',    # '\n' is to end the line there
      centres['Indonesia'].value_counts(), '\n',
      centres['Malta'].value_counts(), '\n',
      centres['India'].value_counts())
      
```

```python
# Creating Contigency table

defective_error_free_table = np.array([[271, 267, 269, 280],[29, 33,31,20]])
defective_error_free_table
```

```python
stats.chi2_contingency(defective_error_free_table)
```

```python
alpha = 0.05
pvalue = 0.2771020991233135


if pvalue < alpha:
  print("We Reject Null hypothesis defective % varies by centre and somewhat/significantly related and there is dependency")
else:
  print("We fail to Reject Null hypothesis because of lack of evidence")
```

So, We fail to Reject Null hypothesis because of lack of evidence and  defective % does not vary by centre and there is no dependency on centres
