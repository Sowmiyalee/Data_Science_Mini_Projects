# **Question 3**

**Sales of products in four different regions is tabulated for males and females. Find if male-female buyer rations are similar across regions.**



*   H0: There is no association or dependency between the gender based buyer rations across regions

*   H1: There is a significant association or dependency between the gender based buyer rations across regions

We are going to conduct a Test of Independence using Chi-Square test with Contingency table since We need to check whether the proportion of any of these samples are different or the same


```python
buyer_ratio= pd.read_csv('/content/BuyerRatio.csv')
buyer_ratio.head()
```

```python
df = buyer_ratio.iloc[:,1:5]
df
```

```python
#Applying Chi-Square contingency table to convert observed value into expected value

stats.chi2_contingency(df)
```

```python
alpha = 0.05
p_value= 0.6603094907091882

if p_value < alpha:
      print("We reject Null hypothesis and say, there is a significant association and dependency in the gender based buying ratio across the regions")
else:
      print(" We fail to reject null hypotheis")
```

We fail to reject Null Hypothesis because of lack evidence. Therefore, there is no association or dependency between male-female buyers rations and are similar across regions. Hence, Independent samples
