```python
#import libraries
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#import data
data = pd.read_csv("/content/forestfires.csv")
data
```

## EDA

```python
data = data.drop(['month','day'],axis=1)
data
```

```python
#label encoder the variables to make small as 1 and large as 0
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

## Model Building - Basic SVM

```python
#Dividing the data into independent and dependent variable
X = data.drop(['size_category'], axis=1)
y = data['size_category']
```

```python
#Splitting the dataset into train and test dataset
from sklearn.model_selection import train_test_split
X_train, X_test ,y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=0)
```

```python
#Feature scaling
#from sklearn.preprocessing import StandardScaler
#sc_X=StandardScaler()
#X_train_fit = sc_X.fit_transform(X_train)
#X_test_fit = sc_X.transform(X_test)
```

```python
#Fitting SVM model
from sklearn.svm import SVC
classifier = SVC() 
classifier.fit(X_train, y_train)
```

### Model prediction

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
y_test.values
```

### Model Accuracy

```python
#Confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
#Accuracy score
from sklearn.metrics import accuracy_score
SVM_ac = accuracy_score(y_pred,y_test)
SVM_ac*100
```

## Model building - Kernel=Linear

```python
from sklearn.svm import SVC
classifier = SVC(kernel='linear')
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
cm = confusion_matrix(y_test,y_pred)
from sklearn.metrics import accuracy_score
Lin_ac = accuracy_score(y_pred,y_test)
Lin_ac*100
```

```python
#with gamma value
classifier5 = SVC (kernel='linear', gamma=0.008, C=1.3)
classifier5.fit(X_train,y_train)
y_pred5 = classifier5.predict(X_test)
lin_ac_g= accuracy_score(y_pred5,y_test)
lin_ac_g*100
```

```python
#with change in penalty parameter accuracy change
classifier5 = SVC (kernel='linear', gamma=0.008, C=1.4)
classifier5.fit(X_train,y_train)
y_pred5 = classifier5.predict(X_test)
lin_ac_g= accuracy_score(y_pred5,y_test)
lin_ac_g*100
```

```python
# when we apply c=1.3 we get accuracy as 97.69,
# when we apply c=1.4 we get accuracy as 98.46.
```

 ## Model building - Kernel=Poly

```python
#without gamma value
classifier3 = SVC (kernel='poly')
classifier3.fit(X_train,y_train)
y_pred3 = classifier3.predict(X_test)
cm = confusion_matrix(y_test,y_pred3)
Pol_ac = accuracy_score(y_pred3,y_test)
Pol_ac*100
```

## Model building - Kernel=RBF

```python
classifier5 = SVC (kernel='rbf', gamma=0.008)
classifier5.fit(X_train,y_train)
y_pred5 = classifier5.predict(X_test)
rbf_ac_g3= accuracy_score(y_pred5,y_test)
rbf_ac_g3*100
```

```python
classifier5 = SVC (kernel='rbf', gamma=0.009)
classifier5.fit(X_train,y_train)
y_pred5 = classifier5.predict(X_test)
rbf_ac_g3= accuracy_score(y_pred5,y_test)
rbf_ac_g3*100
```

```python
# when we apply gamma=0.008 we get accuracy as 74.61,
# when we apply gamma=0.009 we get accuracy as 73.84.
#Accuracy level decreses when we exceed the gamma value above 0.008
```
